```python
print('slope:', lm_m_from_time.coef_)
print('intercept:', lm_m_from_time.intercept_)
```

```python
pred_m_from_time = lm_m_from_time.predict(time)

df_pred_m_from_time = pd.DataFrame({'사용시간':time.flatten(), '실제 이동거리':m.flatten(), '예측 이동거리':pred_m_from_time.flatten()})
df_pred_m_from_time
```
